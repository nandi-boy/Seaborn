
>Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

```python
bins = [2,3,4,5,6,7,8,9,10]
sns.distplot(n['time_spend_company'],bins=bins,color = 'green')
plt.xticks(bins)
plt.show()
```

```python
sns.distplot(n['time_spend_company'],kde=False)
plt.show()
```

```python
sns.distplot(n['time_spend_company'],hist = False)
plt.show()
```

```python
sns.distplot(n['time_spend_company'], bins = bins, rug = True,color='green',hist_kws={'color':'pink' , 'edgecolor':'blue' , 'linewidth':3 , 'alpha': 0.3},kde_kws={'color':'orange' , 'linewidth':2})
plt.show()
```

## Bar Plots

> A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df
```

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df )
plt.show()
```

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,hue = 'sex')
plt.show()
```

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df , hue = 'sex',palette = 'icefire',alpha=0.5)
plt.show()
```

```python
sns.barplot(x = 'fare',y = 'class' , data = titanic_df ,hue = 'sex',palette = 'inferno',orient = 'h')
#orient='h' ->  It tells the plot to treat the data horizontally.
plt.show()
```

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,hue = 'sex',palette = 'gist_rainbow',estimator = np.max)
# estimator=np.max	-> Instead of default mean fare, this shows the maximum fare for each group
plt.show()
```

```python
#confidence interval
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,hue = 'sex',palette = 'gist_rainbow',ci=50,errcolor = 'orange',errwidth = 5,)
plt.show()
```

>ci=50      -> 	Draws 50% confidence interval (default is 95%) — narrower error bars

>errcolor='orange'	->  Sets the color of the error bars to orange

>errwidth=5	 ->  Makes the error bars thicker (5px) for visibility

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,hue = 'sex',palette = 'inferno',saturation = 0.5)
plt.show()
# saturation=0.5 ->	Controls the intensity of the color (on a scale of 0 to 1). 0.5 makes the colors less saturated, i.e., softer appearance.
```

## HeatMap

> * A heatmap in Seaborn is a two-dimensional graphical representation of data where individual values within a matrix are represented by colors. It is a powerful tool for visualizing patterns, correlations, and magnitudes within rectangular datasets.
> * **USE:** Heatmaps are visual tools used to represent data using colors, where the intensity of the color corresponds to the value of the data. They are primarily used to identify patterns, trends, and areas of interest within a dataset, making it easier to understand complex information. Heatmaps are versatile and can be applied in various fields, including website optimization, scientific research, and geographical analysis.  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

* linecolor='yellow'   ->	 Sets the color of those lines to yellow.
* cbar=False   -> 	Hides the color bar legend (normally shown on the right).
* xticklabels=False   ->	Hides the column labels (top of the heatmap)
* yticklabels=False  ->	Hides the row labels (left side of the heatmap)

## Violin Plot

> * A Seaborn violin plot is a data visualization that combines aspects of a box plot and a kernel density estimate (KDE) plot. It is used to display the distribution of quantitative data across different categories of one or more categorical variables. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Strip Plot

> * A strip plot is a type of data visualization that displays the distribution of a numeric variable across different categories. It's essentially a scatter plot where the data points are plotted along a single axis (usually the y-axis) and distributed randomly along the other axis (usually the x-axis). This randomization helps to reduce overplotting and reveal the underlying distribution of the data points within each category. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vr = sns.load_dataset('tips')
vr
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',palette ='rocket_r',linewidth=0.5,edgecolor='green',data=vr)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.stripplot(x='day',y='total_bill',hue='sex',jitter=1,data=vr)     # jitter=1	-> Adds maximum spread to help separate overlapping points.
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',jitter=0,data=vr)     # jitter=1	-> No spread — points may overlap.
plt.show()
```

```python
plt.figure(figsize=(6,10))
sns.stripplot(x='day',y='total_bill',hue='sex',linewidth=0.5,size=10,data=vr)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',linewidth=0.5,size=10,marker='*',data=vr)
plt.show()
```

```python
sns.stripplot(x=vr['total_bill'],color='green',data=vr)
plt.show()
```

```python
sns.stripplot(y=vr['total_bill'],color='green',data=vr)
plt.show()
```

## Box Plot

> **Box Plot:**  Box Plot is a graphical method to visualize data distribution for gaining insights and making informed decisions.
**Elements of Box Plot:**
A box plot gives a five-number summary of a set of data which is-
>
>* Minimum - It is the minimum value in the dataset excluding the outliers.
>
>* First Quartile (Q1) - 25% of the data lies below the First (lower) Quartile.
>
>* Median (Q2) - It is the mid-point of the dataset. Half of the values lie below it and half above.
>
>* Third Quartile (Q3) - 75% of the data lies below the Third (Upper) Quartile.
>
>* Maximum - It is the maximum value in the dataset excluding the outliers.


> * IQR = Q3-Q1
> * Lower Limit = Q1 - 1.5*IQR
> * Upper Limit = Q3 + 1.5*IQR

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rv=sns.load_dataset('tips')
rv
```

```python
sns.boxplot(x='day',y='total_bill',data=rv)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day',y='total_bill',hue='sex',palette='cool',order=['Fri','Sat','Sun','Thur'],showmeans=True,meanprops={'marker':'+','markeredgecolor':'red'},linewidth=3,data=rv)
plt.show()
```

> showmeans=True  ->  Displays the mean value as a special marker inside each box
> 
> meanprops={'marker':'+', 'markeredgecolor':'red'}	Customizes how the mean marker looks: '+' symbol : Red edge color


> sns.set(style="whitegrid")
> * Sets the plot background style to white with grid lines.
> * Makes it easier to read data values across the plot.

```python
sns.set(style="whitegrid")
sns.boxplot(y='day',x='total_bill',hue='sex',palette='cool',order=['Fri','Sat','Sun','Thur'],showmeans=True,meanprops={'marker':'+','markeredgecolor':'red'},linewidth=3,data=rv)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=rv['total_bill'],color='orange')
plt.show()
```

## Multiple Plots (Facet-Grid)

> **FacetGrid:** FacetGrid is a general way of plotting grids based on a function. It helps in visualizing distribution of one variable as well as the relationship between multiple variables. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ns = sns.load_dataset('tips')
ns
```

```python
fg = sns.FacetGrid(ns)
fg.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
fg = sns.FacetGrid(ns,col='day',hue='sex',palette='summer')
fg.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(ns,col='day',hue='sex',palette='Accent')
fg.map(plt.bar,'total_bill','tip').add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(ns,col='day',hue='smoker',palette='Accent')
fg.map(plt.bar,'total_bill','tip').add_legend()
plt.show()
```
